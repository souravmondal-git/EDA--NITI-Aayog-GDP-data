# Categorization using quantile

```python
category_1 = []
category_2 = []
category_3 = []
category_4 = []
for state in states:
    if(gsva.loc[state,'Per Capita GSDP (Rs.)'] <= per_capita.quantile(0.2)):
        category_4.append(state)
    elif((gsva.loc[state,'Per Capita GSDP (Rs.)'] > per_capita.quantile(0.2)) and (gsva.loc[state,'Per Capita GSDP (Rs.)'] <= per_capita.quantile(0.5))):
        category_3.append(state)
    elif((gsva.loc[state,'Per Capita GSDP (Rs.)'] > per_capita.quantile(0.5)) and (gsva.loc[state,'Per Capita GSDP (Rs.)'] <= per_capita.quantile(0.85))):
        category_2.append(state)
    else :
        category_1.append(state)
    
```
